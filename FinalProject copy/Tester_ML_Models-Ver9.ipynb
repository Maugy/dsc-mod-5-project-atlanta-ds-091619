```python
def clean(raw):
    raw = ' '.join(raw.tolist())
    for char in '-.,\n':
        raw = raw.replace(char,' ')
    
    pattern = "([a-zA-Z]+(?:'[a-z]+)?)"
    raw = nltk.regexp_tokenize(raw, pattern)
#     raw = str(raw)
    return raw

def cleanC(raw):
#     raw = ' '.join(raw.tolist())
    for char in '-.,\n':
        raw = raw.replace(char,' ')
    return raw

def C_plus(raw):
    Cplus = re.findall(r'(?i)\bC\+\+(?!\w)', str(raw))
    return Cplus

def C_sharp(raw):
    Csharp = re.findall(r'(?i)\bC\#(?!\w)', str(raw))
    return Csharp

```
